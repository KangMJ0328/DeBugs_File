```python
# df_1이라는 변수에 데이터프레임 복사
df_1 = df.copy()
```

```python
def getlyrics():
  filepath = '/content/drive/MyDrive/프로젝트/Final_Song_File.csv'
  df = pd.read_csv(filepath)
  df_1 = df.copy()
  for i in range(0,16920):
    # df_1['가사']의 2번째 가사 뽑아오기 노래:좋니
    df_1['가사'][i]
    # str 타입을 list로 변환
    from ast import literal_eval
    df_1['가사'][i].split(',')
    # 컬럼명에 가사 주고 인덱스에 가사내용 넣은 후 (구분자)split() ,(콤마)로 구분 짓기
    df= pd.DataFrame({"가사":df_1['가사'][i].split(',')})
    # 가사의 빈도수 확인 
    df.value_counts().to_frame()
    df.value_counts().to_frame()[df.value_counts().to_frame()[0]>=2]
    list_from_df.append(df.value_counts().to_frame()[df.value_counts().to_frame()[0]>=2].index.tolist())
    # del list_from_df[0]
  return list_from_df
```

```python
getlyrics()
```

```python
import pandas as pd
df = pd.DataFrame({'제목':df_title,'가수':df_artist,'가사':list_from_df})
df.to_csv('SongChorus.csv', encoding='UTF-8')
```
