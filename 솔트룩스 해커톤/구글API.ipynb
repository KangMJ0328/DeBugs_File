```python
import googletrans
from googletrans import Translator
 
# print(googletrans.LANGUAGES)
 
text1 = input('한국어를 일본어로 번역합니다 : '  )
 
text2 = input('한국어를 독일어로 번역합니다 : ')
 
text3 = input('한국어를 영어로 번역합니다 : ' )

text4 = input('한국어를 중국어로 번역합니다 : ' )

text5 = input('한국어를 프랑스어로 번역합니다 : ' )

translator = Translator()
 
print(translator.detect(text1))
print(translator.detect(text2))
print(translator.detect(text3))
print(translator.detect(text4))
print(translator.detect(text5))


trans1 = translator.translate(text1, src='ko', dest='ja')
trans2 = translator.translate(text2, src='ko', dest='de')
trans3 = translator.translate(text3, src='ko', dest='en')
trans4 = translator.translate(text4, src='ko', dest='zh-TW')
trans5 = translator.translate(text5, src='ko', dest='fr')
 
print("korean to Japanese: ", trans1.text)
print("Korean to Germany: ", trans2.text)
print("Korean to English: ", trans3.text)
print("Korean to chinese: ", trans4.text)
print("Korean to France: ", trans5.text)
```
