```python
# import tensorflow.compat.v1 as tf
# tf.disable_v2_behavior()
import numpy as np
import pandas as pd

# S: 디코딩 입력의 시작을 나타내는 심볼
# E: 디코딩 출력을 끝을 나타내는 심볼
# P: 현재 배치 데이터의 time step 크기보다 작은 경우 빈 시퀀스를 채우는 심볼
#    예) 현재 배치 데이터의 최대 크기가 4 인 경우
#       word -> ['w', 'o', 'r', 'd']
#       to   -> ['t', 'o', 'P', 'P']
char_arr = [c for c in '[]!+,.0123456789?ABCDFILNOPSTUVcilmpsuv~…○、。々〜ぁあいぅうぇえおかがきぎくぐけげこごさざしじすずせぜそぞただちっつてでとどなにねのはばぱひびふぶぷへべほまみむめもゃやゆょよらりるれろわをんァアィイゥウェエォオカガキギクグケゲコゴサザシジスズセゼソタダチッツテデトドナニネノハバパヒビピフブプベペホボポマミムメモャュユョヨラリルレロワン・ー一丁万丈三上下不両並中主久乗九乳亀了予事二互五交人今介仕他付代以仮件任企休会伝伺似低住体何余作使供依価便促保信修俺倉個倒借値健側備傾働像僕億償元兄先光免入全公共具内円再冗写冬冷凍凡処出分切列初別到制削前割力加助労効勉動務勧包化十千午半卓南単原参反収取受口古可台右号司合同名向君吸告周味呼命品員唾商問喫営器四回団困国園土圧在地坊型埋基堂報場境売変夏夕外多夢大天太夫失契奥女好姉始委婚嬉子存学宅安完定実客室宮家容寂寄寒寝寧対寿小少就尾局屈届屋属山島州左差布希席帯帰帳常年幸広序店度座庫康弊式引強当形影待後復徴心必忘忙応念怖思急性悩悪情惑惜意愛感態慣慶懸成我戦戻所手打払扱承技投抜押担持振掃掛探接提揚換携撮支改教散数整文料斬断新方旅日早明映時晩景暗暢暮曜曲更書替最月有服望朝期木未末本材来杯東板析枚果栗格案棚椅検業楽構様横機欄次欲歌止正歩歯残段殺毒比気水氷求汗決沢沫況泊法注泳洋洗洞活派流浅浦浮海消液涼混添清済渡測港満準濃点無焼煙煽燃片物特犬状率玉玩現球理璧環生産用由申町画留番異疲病痩癒発登白百的盤目直相看真眠着知短石研確磁磨社祈神福私移程税種積究空窮立筋策算節簡米粉粧精系約納級素索紫細紹終組経結絡給統絶続維緒線締縁縄縦繊繋置署美翻考者聞職肉肌育背能脂脱腰腹自致興船良色花芸若苦英草荷菓落蔚薄薬融血行術街衣表袋装裕補製西要見覧親観角解言計討設訳証評試詫詰話該詳認語誤説読課調談請論諾謝謡議豚負財販買貸費資質購走起越足路跳蹴身車軽載輸辛農込近迷追退送途通速造連週進遅遊運過道達違遠遣適遮選部都配酎酒酢重野量金釜鉄鋼録鍵長閉開間関防降限院除陵階隔際障集難雨電青面靴韓音響頃項領頭頻頼題額顔願類飛食飯飲飼館香駄駅騒験騙驚高鬱鮮鴨麦麺黄黒龍가각간갈감갑갔강갖같개객거걱건걷걸검겁것게겠겨격견결겸겼경계고곤골곳공과관광괜굉교구국군궁권귀그근글금급기긴길김깃까깐깔깜깝꺼껏께꼬꼭꽃꽉꽤꾀꾸꿔뀔끔끝끼낌나난날남납났낭내낼냈냉냐냥너넌널넓넘넣네넨녀녁년녕노논놀농높놓놨뇨누눌눠뉴늄느는늘능늦니닌님닙닝다단닫달닮담답당대댕더던데델도독돈돌동돼됐되된될됩두둔둘뒤뒷드득든듣들듬듭듯등디딜따딱땀때땐떠떡떤떻또뚱뛰라락란랄람랍랐랑래랙랜램랫랬량러런럴럼럽렁렇레렉렌렛려력련렬렴렵렸로록론롭롱뢰료룡루룸류르른를름릅릉리릭린릴림립링마막만많말맑맘맙맛망맞맡매맥맨맵머먹먼멀메멜며면명몇모목몬몰못무묵문물뭐뭔뭘미민밀밌바박반받발밝밤밥방배백버벅번벌범법벗베벤벨벽변별병보복볶본볼봅봉봐봤뵐뵙부분불붙브블비빈빌빔빠빨빴빵빼뽑쁘쁜사삭산살상새색샘생서석섞선설섬섯성세셔션셨소속손송쇠쇼수순술쉐쉬쉽슈스슨습슷승시식신실싫심십싱싶싸싼쌀써썬쎄쏘쓰쓸씀씨씩씬씻아악안앉않알압았앞애액앤앱야약얇양얘어억언얻얼엄업없엇었에엔엘엠여역연열였영예오온올옮옵와완왔왜외왼요용우운울워원월웨웹위유육율으은을음응의이인일읽잃임입있잊자작잔잖잘잠잡장재저적전절젊점접정제젤져졌조좀종좋좌죄죠주죽준줄중줘쥴즈즐즘증지직진질짐집징짜짧짱째쩌쩔쪽쭉쭤쯤찌찍찟차착찬찮참창찾채책챙처천철첨첫청체쳇쳐초촉총촬최추축출춤충춰취츄츠측층치친칠침칭카칵칸캉캐캔캘캠커컨컴컷케켓켜코콘콜콤콩퀴크클큼키킨킬킷킹타탁탈탕태택터턴테텐텔템토톤통퇴투튀트특튼티틴틸팀팁팅파판팔팜패팩펄페편포폰폴폼표풀품프플피필핏핑하학한할함합핫항해핸했행향허험혀현협형호혹혼홀홈홍화확환활황회획효후훨휠휴흘흠흡희흰히힐힘']
num_dic = {n: i for i, n in enumerate(char_arr)}
dic_len = len(num_dic)

# 영어를 한글로 번역하기 위한 학습 데이터
seq_data 

def make_batch(seq_data):
    input_batch = []
    output_batch = []
    target_batch = []

    for seq in seq_data:
        # 인코더 셀의 입력값. 입력단어의 글자들을 한글자씩 떼어 배열로 만든다.
        input = [num_dic[n] for n in seq[0]]
        # 디코더 셀의 입력값. 시작을 나타내는 S 심볼을 맨 앞에 붙여준다.
        output = [num_dic[n] for n in ('S' + seq[1])]
        # 학습을 위해 비교할 디코더 셀의 출력값. 끝나는 것을 알려주기 위해 마지막에 E 를 붙인다.
        target = [num_dic[n] for n in (seq[1] + 'E')]

        input_batch.append(np.eye(dic_len)[input])
        output_batch.append(np.eye(dic_len)[output])
        # 출력값만 one-hot 인코딩이 아님 (sparse_softmax_cross_entropy_with_logits 사용)
        target_batch.append(target)

    return input_batch, output_batch, target_batch


#########
# 옵션 설정
######
learning_rate = 0.01
n_hidden = 128
total_epoch = 100
# 입력과 출력의 형태가 one-hot 인코딩으로 같으므로 크기도 같다.
n_class = n_input = dic_len


#########
# 신경망 모델 구성
######
# Seq2Seq 모델은 인코더의 입력과 디코더의 입력의 형식이 같다.
# [batch size, time steps, input size]
enc_input = tf.placeholder(tf.float32, [None, None, n_input])
dec_input = tf.placeholder(tf.float32, [None, None, n_input])
# [batch size, time steps]
targets = tf.placeholder(tf.int64, [None, None])


# 인코더 셀을 구성한다.
with tf.variable_scope('encode'):
    enc_cell = tf.nn.rnn_cell.BasicRNNCell(n_hidden)
    enc_cell = tf.nn.rnn_cell.DropoutWrapper(enc_cell, output_keep_prob=0.5)

    outputs, enc_states = tf.nn.dynamic_rnn(enc_cell, enc_input,
                                            dtype=tf.float32)

# 디코더 셀을 구성한다.
with tf.variable_scope('decode'):
    dec_cell = tf.nn.rnn_cell.BasicRNNCell(n_hidden)
    dec_cell = tf.nn.rnn_cell.DropoutWrapper(dec_cell, output_keep_prob=0.5)

    # Seq2Seq 모델은 인코더 셀의 최종 상태값을
    # 디코더 셀의 초기 상태값으로 넣어주는 것이 핵심.
    outputs, dec_states = tf.nn.dynamic_rnn(dec_cell, dec_input,
                                            initial_state=enc_states,
                                            dtype=tf.float32)


model = tf.layers.dense(outputs, n_class, activation=None)


cost = tf.reduce_mean(
            tf.nn.sparse_softmax_cross_entropy_with_logits(
                logits=model, labels=targets))

optimizer = tf.train.AdamOptimizer(learning_rate).minimize(cost)


#########
# 신경망 모델 학습
######
sess = tf.Session()
sess.run(tf.global_variables_initializer())

input_batch, output_batch, target_batch = make_batch(seq_data)

for epoch in range(total_epoch):
    _, loss = sess.run([optimizer, cost],
                       feed_dict={enc_input: input_batch,
                                  dec_input: output_batch,
                                  targets: target_batch})

    print('Epoch:', '%04d' % (epoch + 1),
          'cost =', '{:.6f}'.format(loss))

print('최적화 완료!')


#########
# 번역 테스트
######
# 단어를 입력받아 번역 단어를 예측하고 디코딩하는 함수
def translate(word):
    # 이 모델은 입력값과 출력값 데이터로 [영어단어, 한글단어] 사용하지만,
    # 예측시에는 한글단어를 알지 못하므로, 디코더의 입출력값을 의미 없는 값인 P 값으로 채운다.
    # ['word', 'PPPP']
    seq_data = [word, 'P' * len(word)]

    input_batch, output_batch, target_batch = make_batch([seq_data])

    # 결과가 [batch size, time step, input] 으로 나오기 때문에,
    # 2번째 차원인 input 차원을 argmax 로 취해 가장 확률이 높은 글자를 예측 값으로 만든다.
    prediction = tf.argmax(model, 2)

    result = sess.run(prediction,
                      feed_dict={enc_input: input_batch,
                                 dec_input: output_batch,
                                 targets: target_batch})

    # 결과 값인 숫자의 인덱스에 해당하는 글자를 가져와 글자 배열을 만든다.
    decoded = [char_arr[i] for i in result[0]]

    # 출력의 끝을 의미하는 'E' 이후의 글자들을 제거하고 문자열로 만든다.
    end = decoded.index('E')
    translated = ''.join(decoded[:end])

    return translated


```

```python
print('\n=== 번역 테스트 ===')

print('word ->', translate('word'))
print('love ->', translate('love'))
print('kiss ->', translate('kiss'))
```
