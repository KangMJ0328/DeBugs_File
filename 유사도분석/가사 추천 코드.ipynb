```python
import pandas as pd
filepah = '/content/drive/MyDrive/프로젝트/노래 csv파일/SongChorus.csv'
df = pd.read_csv(filepah)
# Unnamde: 0 삭제
df=df.drop('Unnamed: 0',axis=1)
# 공백 행 삭제
# df = df[df.가사 != '[]']

# df.iloc[:,[0,1,2]]
df
```

```python
df.duplicated()
```

```python
# drop_duplicates 함수로 중복된 값 제거 후 ignore_index 로 인덱스 재정렬)
df=df.drop_duplicates(keep='first',ignore_index=True)
```

```python
df[df['제목']=='사계 (Four Seasons)']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
from sklearn.metrics.pairwise import cosine_similarity

count_vect = CountVectorizer(min_df=0,ngram_range=(1,3)).fit(df['가사'])
count_vect.vocabulary_

lyrics_matrix = count_vect.fit_transform(df['가사'])

lyrics_similarity = cosine_similarity(lyrics_matrix,lyrics_matrix)

lyrics_similarity_sorted_idx = lyrics_similarity.argsort()[:,::-1]
```

```python
def name():
  # 사용자가 원하는 노래 제목 입력 
  song_name = input("노래 제목 : ")
  # 사용자가 입력한 노래 제목을 title_song에 저장, (사용자가 입력한 제목의 인덱스 값을 찾기 위한 수단)
  title_song = df[df['제목']== song_name]
  # 노래중에 중복된 노래가 있을수 있으므로 첫번째로 뜬 값으로 사용
  title_song = title_song
  # 노래의 인덱스값 뽑아내는 함수
  song_index = title_song.index.values
  # 코사인유사도 분석 후 노래와 가장 유사한 노래들 10가지 뽑아오기
  iloc = lyrics_similarity_sorted_idx[song_index,:11]
  # 뽑아온 인덱스 값들 정렬 및 저장
  quiz = iloc.reshape(-1)
  # 검색 노래 제외
  quiz = quiz[quiz!=song_index]
  # 가중치 점수를 이용한 탑10 노래 추천
  return df.iloc[quiz][:11]
name()
```
